```python

import os, sys

# путь к родительской папке, в которой лежат и prompt_test, и текущий проект
PROJECT_ROOT = os.path.abspath("..")
if PROJECT_ROOT not in sys.path:
    sys.path.append(PROJECT_ROOT)

print(PROJECT_ROOT)
from contextlib import contextmanager

@contextmanager
def patch_cfg(cfg, **overrides):
    """
    overrides: ключи вида "data.max_length"="512", "model.pretrained_name"="...", "train.output_dir"="..."
    """
    old = {}
    try:
        # применяем overrides
        for path, new_value in overrides.items():
            obj = cfg
            parts = path.split(".")
            for p in parts[:-1]:
                obj = getattr(obj, p)
            attr = parts[-1]
            old[path] = getattr(obj, attr)
            setattr(obj, attr, new_value)
        yield cfg
    finally:
        # откатываем
        for path, old_value in old.items():
            obj = cfg
            parts = path.split(".")
            for p in parts[:-1]:
                obj = getattr(obj, p)
            setattr(obj, parts[-1], old_value)

from train import train
SWEEP = [
#     {
#         "model": "MoritzLaurer/DeBERTa-v3-large-mnli-fever-anli-ling-wanli",
#         "lengths": [128, 256, 512],
#         "tag": "deberta512",
#     },
#     {
#         "model": "Tochka-AI/ruRoPEBert-e5-base-2k",
#         "lengths": [256, 512, 1024, 1536, 2048],
#         "tag": "rope2k",
#     },
    {
        "model": "deepvk/RuModernBERT-base",  # пример
        "lengths": [4096, 8192], #512, 1024, 2048, 
        "tag": "modern8k",
    },
]

TRACK = [
    "relevance_f1",
    "utilization_f1",
    "adherence_f1",
]
from pathlib import Path
from config import cfg
from data import safe_model_tag
import pickle
BASE = Path().resolve()

results = []
results_all = []
for item in SWEEP:
    model_name = item["model"]
    lengths = item["lengths"]
    tag = item.get("tag", model_name)

    for L in lengths:
        out_dir = str(BASE / "sweeps" / safe_model_tag(model_name) / f"len_{L}")
        print(f"\n=== RUN model={model_name} max_length={L} out={out_dir} ===")
        with patch_cfg(
            cfg,
            **{
                "model.pretrained_name": model_name,
                "data.max_length": L,
                "train.output_dir": out_dir,
            }
        ):
            res = train()

        valm = res.get("class_metrics", {}) if isinstance(res, dict) else {}
        row = {
            "tag": tag,
            "model": model_name,
            "max_length": L,
            "output_dir": res.get("output_dir") if isinstance(res, dict) else None,
        }
        for k in TRACK:
            row[k] = float(valm.get(k, float("nan")))
        results.append(row)
        results_all.append(valm)
        with open('tmp4.pkl', 'wb') as f:
            pickle.dump(results_all, f )

print("\nDone. Total runs:", len(results))
```

```python
import pickle
from pathlib import Path
import pandas as pd


SWEEP = [
    {
        "model": "MoritzLaurer/DeBERTa-v3-large-mnli-fever-anli-ling-wanli",
        "lengths": [128, 256, 512],
        "tag": "deberta512",
    },
    {
        "model": "Tochka-AI/ruRoPEBert-e5-base-2k",
        "lengths": [256, 512, 1024, 1536, 2048],
        "tag": "rope2k",
    },
    {
        "model": "deepvk/RuModernBERT-base",
        "lengths": [512, 1024, 2048, 4096, 8192],
        "tag": "modern8k",
    },
]


PKL_DIR = Path(".")           
PKL_FILES = sorted(PKL_DIR.glob("tmp*.pkl")) 

print("Found PKLs:", [f.name for f in PKL_FILES])

run_plan = []
for item in SWEEP:
    for L in item["lengths"]:
        run_plan.append({
            "tag": item.get("tag"),
            "model": item["model"],
            "max_length": L,
        })

print("Planned runs:", len(run_plan))
print("First planned runs:", run_plan[:5])

def load_metrics_list(p: Path):
    with p.open("rb") as f:
        obj = pickle.load(f)
    if not isinstance(obj, list):
        raise TypeError(f"{p}: expected list, got {type(obj)}")
    # ожидаем list[dict]
    return obj


all_lists = [(p, load_metrics_list(p)) for p in PKL_FILES if p.exists()]
if not all_lists:
    raise FileNotFoundError("No tmp*.pkl found in this folder")
metrics_list = []
for i in all_lists:
    metrics_list.extend(i[1])
# best_path, metrics_list = max(all_lists, key=lambda x: len(x[1]))
# print("Using:", best_path.name, "len =", len(metrics_list))

# --- сопоставляем по индексу i -> run_plan[i] ---
rows = []
n = min(len(metrics_list), len(run_plan))
print(len(metrics_list))
for i in range(n):
    m = metrics_list[i]
    rp = run_plan[i]
    row = {"run_index": i, **rp}


    if isinstance(m, dict):
        row.update(m)
    rows.append(row)

df = pd.DataFrame(rows).sort_values(["tag","model","max_length"]).reset_index(drop=True)

cols = ["run_index","tag","model","max_length",'relevance_precision', 'relevance_recall', 'relevance_f1', 'utilization_precision', 'utilization_recall', 'utilization_f1', 'adherence_precision', 'adherence_recall', 'adherence_f1']
df[[c for c in cols if c in df.columns]].head(20)
```

```python
import matplotlib.pyplot as plt

metrics = ["relevance_f1", "utilization_f1", "adherence_f1"]

group_col = "tag" if "tag" in df.columns and df["tag"].notna().any() else "model"

for metric in metrics:
    if metric not in df.columns:
        print("Missing metric:", metric)
        continue

    plt.figure()
    for name, g in df.groupby(group_col):
        g = g.sort_values("max_length")
        plt.plot(g["max_length"], g[metric], marker="o", label=str(name))

    plt.xlabel("max_length")
    plt.ylabel(metric)
    plt.title(f"{metric} vs max_length ({group_col})")
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()
```

```python
import matplotlib.pyplot as plt
from pathlib import Path
import re

metrics = ["relevance_f1", "utilization_f1", "adherence_f1"]
group_col = "tag" if "tag" in df.columns and df["tag"].notna().any() else "model"

OUT_DIR = Path("plots")
OUT_DIR.mkdir(parents=True, exist_ok=True)

def slug(s: str) -> str:
    s = str(s)
    s = re.sub(r"[^a-zA-Z0-9._-]+", "_", s)
    return s.strip("_")[:120]

fig, axes = plt.subplots(1, 3, figsize=(16, 4), sharex=True)

for ax, metric in zip(axes, metrics):
    if metric not in df.columns:
        ax.set_title(f"{metric} (missing)")
        ax.axis("off")
        continue

    for name, g in df.groupby(group_col):
        g = g.sort_values("max_length")
        ax.plot(g["max_length"], g[metric], marker="o", label=str(name))

    ax.set_xlabel("max_length")
    ax.set_ylabel(metric)
    ax.set_title(metric)
    ax.grid(True, alpha=0.3)

handles, labels = axes[0].get_legend_handles_labels()
if handles:
    fig.legend(handles, labels, loc="upper center", ncol=min(len(labels), 4), bbox_to_anchor=(0.5, 1.15))

fig.suptitle(f"Metrics vs max_length (grouped by {group_col})", y=1.25)
fig.tight_layout()

fname = OUT_DIR / f"metrics_1x3__by_{slug(group_col)}.png"
fig.savefig(fname, dpi=200, bbox_inches="tight")
plt.close(fig)

print("Saved:", fname)
```
